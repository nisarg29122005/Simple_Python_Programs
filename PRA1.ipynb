```
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

# Example usage:
number = 5
print(f"The factorial of {number} is {factorial(number)}")

number = 0
print(f"The factorial of {number} is {factorial(number)}")

number = -3
print(f"The factorial of {number} is {factorial(number)}")
```

```
def display_multiplication_table(number, limit=10):
    print(f"Multiplication Table for {number}:")
    for i in range(1, limit + 1):
        print(f"{number} x {i} = {number * i}")

# Example usage:
num_to_display = 7
display_multiplication_table(num_to_display)

print("\n") # Add a newline for better separation

num_to_display_2 = 12
display_multiplication_table(num_to_display_2, limit=5)
```

```
def check_odd_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Example usage:
print(check_odd_even(4))
print(check_odd_even(7))
print(check_odd_even(0))
print(check_odd_even(-5))
```

```
def is_palindrome(number):
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Compare the original string with its reversed version
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# Example usage:
print(is_palindrome(121))
print(is_palindrome(12321))
print(is_palindrome(123))
print(is_palindrome(0))
print(is_palindrome(1))
print(is_palindrome(-121)) # Note: Palindrome check usually applies to positive numbers or absolute value if specified.
```

```
def expert_system():
    print("=== Medical Diagnosis Expert System ===")
    fever = input("Do you have fever? (yes/no): ").lower()
    cough = input("Do you have cough? (yes/no): ").lower()
    headache = input("Do you have headache? (yes/no): ").lower()
    sneezing = input("Do you have sneezing? (yes/no): ").lower()

    # Inference Engine (Rules)
    if fever == "yes" and cough == "yes" and headache == "yes":
        diagnosis = "You may have FLU."
    elif sneezing == "yes" and cough == "yes" and fever == "no":
        diagnosis = "You may have COMMON COLD."
    elif fever == "yes" and headache == "yes":
        diagnosis = "You may have VIRAL FEVER."
    else:
        diagnosis = "Condition unclear. Please consult a doctor."

    print("\nDiagnosis Result:")
    print(diagnosis)

# Run the expert system
expert_system()
```
